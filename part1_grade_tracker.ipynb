```python
# Task 1 — Data Parsing & Profile Cleaning

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:

    # strip whitespace and convert to title case
    clean_name = student["name"].strip().title()

    # convert roll from string to integer
    clean_roll = int(student["roll"])

    # split marks on ", " and convert each piece to int
    clean_marks = [int(m) for m in student["marks_str"].split(", ")]

    cleaned = {
        "name":  clean_name,
        "roll":  clean_roll,
        "marks": clean_marks,
    }
    cleaned_students.append(cleaned)

    # isalpha() returns True only if all characters in the word are letters
    all_words_alpha = all(word.isalpha() for word in clean_name.split())
    validity = "✓ Valid name" if all_words_alpha else "✗ Invalid name"

    # print profile card for each student
    print("================================")
    print(f"Student : {clean_name}   {validity}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks   : {clean_marks}")
    print("================================")
    print()

# find roll 103 and print name in upper and lower case
for s in cleaned_students:
    if s["roll"] == 103:
        print(f"Roll 103 in ALL CAPS  : {s['name'].upper()}")
        print(f"Roll 103 in lowercase : {s['name'].lower()}")
        break
```

```python
# Task 2 — Marks Analysis Using Loops & Conditionals

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# Part 1 — print each subject with its grade

print("Results for:", student_name)
print("─" * 35)

for i in range(len(subjects)):
    m = marks[i]

    # checking which range the mark falls in
    if m >= 90:
        grade = "A+"
    elif m >= 80:
        grade = "A"
    elif m >= 70:
        grade = "B"
    elif m >= 60:
        grade = "C"
    else:
        grade = "F"

    print("Subject: " + subjects[i] + "  |  Marks: " + str(m) + "  |  Grade: " + grade)

print("─" * 35)
```

```python
# Part 2 — Total, Average, Highest, Lowest

# calculating total by looping through marks
total = 0
for m in marks:
    total = total + m

average = round(total / len(marks), 2)

# finding highest scoring subject manually
highest_marks   = marks[0]
highest_subject = subjects[0]

for i in range(len(marks)):
    if marks[i] > highest_marks:
        highest_marks   = marks[i]
        highest_subject = subjects[i]

# finding lowest scoring subject manually
lowest_marks   = marks[0]
lowest_subject = subjects[0]

for i in range(len(marks)):
    if marks[i] < lowest_marks:
        lowest_marks   = marks[i]
        lowest_subject = subjects[i]

print("Total Marks       : " + str(total))
print("Average Marks     : " + str(average))
print("Highest Subject   : " + highest_subject + " (" + str(highest_marks) + ")")
print("Lowest Subject    : " + lowest_subject  + " (" + str(lowest_marks)  + ")")
print("─" * 35)
```

```python
# Part 3 — while loop marks entry system

new_subjects_count = 0

while True:
    subject_input = input("Enter subject name (or 'done' to stop): ")

    if subject_input.lower() == "done":
        break

    marks_input = input("Enter marks for " + subject_input + " (0-100): ")

    try:
        marks_value = int(marks_input)

        if marks_value < 0 or marks_value > 100:
            print("⚠ Warning: Marks must be between 0 and 100. Not added.\n")
            continue

        subjects.append(subject_input)
        marks.append(marks_value)
        new_subjects_count = new_subjects_count + 1
        print("Added successfully\n")

    except ValueError:
        print("⚠ Warning: That's not a valid number. Not added.\n")

# calculate updated average after loop
total_updated = 0
for m in marks:
    total_updated = total_updated + m

updated_avg = round(total_updated / len(marks), 2)

print()
print("New subjects added: " + str(new_subjects_count))
print("Updated Average (all subjects): " + str(updated_avg))
```

```python
# Task 3 — Class Performance Summary

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

results = []

# calculate average and pass/fail status for each student
for student in class_data:
    name  = student[0]
    marks = student[1]

    total = 0
    for m in marks:
        total = total + m

    average = round(total / 5, 2)

    if average >= 60:
        status = "Pass"
    else:
        status = "Fail"

    results.append((name, average, status))
```

```python
# print the class report table

print("Name              | Average | Status")
print("----------------------------------------")

for r in results:
    print(f"{r[0]:<18}| {r[1]:^7} | {r[2]}")

print("----------------------------------------")
```

```python
# count pass and fail, find topper, find class average

passed = 0
failed = 0

topper_name = ""
topper_avg  = 0

all_averages = []

for r in results:
    if r[2] == "Pass":
        passed = passed + 1
    else:
        failed = failed + 1

    # check if this student has higher average than current topper
    if r[1] > topper_avg:
        topper_avg  = r[1]
        topper_name = r[0]

    all_averages.append(r[1])

# class average is average of all students averages
class_avg_total = 0
for a in all_averages:
    class_avg_total = class_avg_total + a

class_avg = round(class_avg_total / len(all_averages), 2)

print("Students Passed : " + str(passed))
print("Students Failed : " + str(failed))
print("Class Topper    : " + topper_name + " (" + str(topper_avg) + ")")
print("Class Average   : " + str(class_avg))
```

```python
# Task 4 — String Manipulation Utility

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# step 1 - strip whitespace first, use this for everything below
clean_essay = essay.strip()
print("Step 1 - Cleaned Essay:")
print(clean_essay)
print()

# step 2 - convert to title case
print("Step 2 - Title Case:")
print(clean_essay.title())
print()

# step 3 - count how many times "python" appears
# clean_essay is already lowercase so .count("python") catches everything
count = clean_essay.count("python")
print("Step 3 - Python word count:")
print(count)
print()

# step 4 - replace "python" with "Python 🐍"
replaced = clean_essay.replace("python", "Python 🐍")
print("Step 4 - After Replace:")
print(replaced)
print()

# step 5 - split into sentences by ". "
sentences = clean_essay.split(". ")
print("Step 5 - Sentences List:")
print(sentences)
print()

# step 6 - print each sentence numbered
# add "." at end if it doesn't already have one
print("Step 6 - Numbered Sentences:")
number = 1
for sentence in sentences:
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(str(number) + ". " + sentence)
    number = number + 1
```
